```julia
using JWAS,CSV,DataFrames,JWAS.misc,DelimitedFiles
```

```julia
phenotypes = CSV.read("JWAS_Phenotype.csv",delim = ',',header=true,missingstrings=".")
dropmissing!(phenotypes,[:ExpPen,:ExpPar,:Sex,:BegAge,:Dam,:WUR,:BegWt,:VL,:WG])
```

```julia
VL_model_equations = "VL = intercept + ExpPen + ExpPar + Sex + BegAge + BegWt + Dam"
VL_R=22.5790
VL_model=build_model(VL_model_equations,VL_R);
```

```julia
set_covariate(VL_model,"BegAge BegWt")
```

```julia
VL_G=3.92479 
set_random(VL_model,"Dam",VL_G)
```

```julia
VL_G1=1.68317 
set_random(VL_model,"ExpPen",VL_G1)
```

```julia
VL_G3=19.5378
add_genotypes(VL_model,"JWAS_1421pigsPHGC1-8_52410SNP.csv",VL_G3,separator=',')
```

```julia
#VLoutC=runMCMC(VL_model,phenotypes,methods="BayesC",estimatePi=true,chain_length=50000,burnin = 5000,output_samples_frequency=100)
```

```julia
VLoutB=runMCMC(VL_model,phenotypes,methods="BayesB",Pi=0.99,estimateScale=true, output_heritability=true,chain_length=50000,burnin = 5000,output_samples_frequency=100)
```

```julia
VLoutB=runMCMC(VL_model,phenotypes,methods="BayesB",Pi=0.99,output_heritability=true,chain_length=50000,burnin = 5000,output_samples_frequency=100)
```

```julia
keys(VLoutB)
```

```julia
VL_out,VL_winVarProps = GWAS("MCMC_samples_marker_effects_VL.txt","JWAS_Map_PHGC1-8_52410SNP.txt",VL_model;header=true,window_size="1 Mb",threshold=0.001,output_winVarProps=true)
```

```julia
writedlm( "MCMC_samples_1MbwindowVarProps_Haplotype_VL-BayesB-pi0.99_noWUR-04032019.txt", VL_winVarProps,'\t')
writedlm( "MCMC_samples_1Mbwindow_out_Haplotype_VL-BayesB-pi0.99_noWUR-04032019.txt", VL_out,'\t')
```
